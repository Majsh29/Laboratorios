```python
tu_nombre = "jian"
tu_apellido = "sanchez"
```

```python
# 2. El Registro Sucio
registro_sucio = " ##[ID-9875]__Monto:S/.4500__Estado:PENDIENTE## "
```

```python
Texto_limpio = registro_sucio.strip(" #")
```

```python
texto_minuscula = Texto_limpio.lower()
texto_minuscula
```

```python
texto_reemplazado = texto_minuscula.replace("pendiente","procesado")
texto_reemplazado
```

```python
Lista = texto_reemplazado.split("__")
Lista
```

```python
parte_monto = Lista[1]
print(parte_monto)
```

```python
indice_m = parte_monto.index("m")
indice_m
```

```python
print(f"Reporte generado por {tu_nombre} {tu_apellido}. El texto final limpio y dividido es: {Lista}. La 'm' de monto se encontró en la posición: {indice_m}")
```
