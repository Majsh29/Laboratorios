```python
log_crudo = "$$[ESTADO:ok]||trx_id:9934||monto:S/.1250.50||alerta:falso$$ "
```

```python
Limpio = log_crudo.strip(" $")
Limpio
```

```python
Mayuscula = Limpio.upper()
Mayuscula
```

```python
Lista_Campos = Mayuscula.split("||")
print(Lista_Campos)
```

```python
Parte_Monto_Completa = Lista_Campos[-2]
print(Parte_Monto_Completa)
```

```python
Monto_str = Parte_Monto_Completa[len("MONTO:S/."):]
print(Monto_str)
```

```python
Monto_Transaccion = float(Monto_str)
print(Monto_Transaccion)
```

```python
umbrales_auditoria = (500.0, 2000.0)
```

```python
if umbrales_auditoria[1] < Monto_Transaccion:
 print("ALERTA ROJA: Transacción bloqueada. Supera los $2000." )
elif umbrales_auditoria[1] >= Monto_Transaccion > umbrales_auditoria[0]:
 print("ALERTA AMARILLA: Transacción enviada a revisión manual." )
else:
 print("VERDE: Transacción procesada automáticamente." )
```
