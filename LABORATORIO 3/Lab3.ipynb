```python
respuesta_api = [
    {"trx_id": "T001", "monto_usd": 150.00, "estado": "Completado"},
    {"trx_id": "T002", "monto_usd": "Error_Red", "estado": "Fallido"},
    {"trx_id": "T003", "monto_usd": 320.50, "estado": "Completado"},
    {"trx_id": "T004", "monto_usd": 45.00, "estado": "Completado"}
]
```

```python
def extraer_montos(lista_datos):
    """
    Recorre una lista de diccionarios y extrae únicamente los valores de 'monto_usd'.
    """
    lista_montos = []
    for registro in lista_datos:
        # Extraemos el valor asociado a la llave "monto_usd" en cada ciclo
        monto = registro["monto_usd"]
        lista_montos.append(monto)
    return lista_montos
```

```python
# Prueba de la Parte 1
montos_extraidos = extraer_montos(respuesta_api)
print("1. Montos extraídos de la API:", montos_extraidos)
```

```python
def convertir_a_soles(monto):
    """
    Convierte un monto en dólares a soles (tipo de cambio 3.8).
    Incluye un mecanismo de seguridad para evitar colapsos por datos corruptos.
    """
    # IMPLEMENTACIÓN DEL TRY/EXCEPT:
    # Se utiliza este bloque porque sabemos que la API envía ocasionalmente texto
    # en lugar de números (ej. "Error_Red"). Si intentamos multiplicar un texto,
    # Python lanzará un TypeError y detendrá todo el sistema. El try/except
    # 'atrapa' ese error, reporta el problema y permite que el código siga fluyendo.
    try:
        conversion = monto * 3.8
        return round(conversion, 2)
    except Exception as e:
        print(f"\n⚠️ Alerta: Dato corrupto detectado. Detalle: [{e}]")
        return 0
```

```python
convertir_a_soles(montos_extraidos)
```

```python
# Lista proporcionada por el área de Finanzas
montos_limpios = [570.0, 0, 1217.9, 171.0, 850.0, 90.0]

# List Comprehension: [resultado FOR elemento IN lista IF condición]
# Multiplicamos por 0.18 para sacar el impuesto, redondeamos a 2 decimales,
# y aplicamos el filtro estricto de > 500.0
impuestos_grandes = [round(monto * 0.18, 2) for monto in montos_limpios if monto > 500.0]

print("\n3. Impuestos de transacciones grandes:", impuestos_grandes)
```

```python
# Calculamos la suma total de la lista filtrada
suma_total_impuestos = sum(impuestos_grandes)

# Reporte final utilizando f-string y caracteres de escape (\n) para saltos de línea
reporte_final = f"""
======================================
         REPORTE DE AUDITORÍA
======================================
Total de impuestos extraídos: {len(impuestos_grandes)} registros.
Suma total de impuestos a retener: S/ {suma_total_impuestos}
======================================\nOperación finalizada sin interrupciones del sistema.
"""
```

```python
print(reporte_final)
```
